### Correlation Summary

14. Create the correlation summary

```python
# Create a comprehensive correlation matrix
correlation_matrix = df.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, mask=mask, fmt='.3f', cbar_kws={"shrink": .8})
plt.title('Correlation Matrix (Lower Triangle)', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

15. Create a correlation matrix

```python
# Create a more detailed correlation matrix with p-values
def correlation_matrix_with_pvalues(df):
    from scipy.stats import pearsonr
    cols = df.columns
    n = len(cols)
    p_matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if i == j:
                p_matrix[i, j] = 0
            else:
                _, p_value = pearsonr(df[cols[i]], df[cols[j]])
                p_matrix[i, j] = p_value

    # Create a formatted table
    corr_df = df.corr().round(3)
    p_df = pd.DataFrame(p_matrix, columns=cols, index=cols).round(4)

    # Combine correlation and p-values
    result = corr_df.astype(str) + " (p=" + p_df.astype(str) + ")"
    np.fill_diagonal(result.values, "-")

    return result

# Display the detailed correlation matrix
corr_summary = correlation_matrix_with_pvalues(df)
print("Correlation Matrix with P-values:")
print(corr_summary)
```

## Building Different Regression Models

Find the best regression model

16. Set independent and dependent variables

```python
# Define independent variables (X) and dependent variable (y)
X = df[['Miles Traveled', 'Number of Deliveries', 'Gas Price']]
y = df['Travel Time']

# Add constant for statsmodels
X_const = sm.add_constant(X)

print("Independent variables shape:", X.shape)
print("Dependent variable shape:", y.shape)
```

17. Build and train the model

```python
# Build and train the model using statsmodels
model = sm.OLS(y, X_const).fit()
```

18. Display the statistical summary

```python
# Display detailed regression results
print(model.summary())

# Calculate VIF to check for multicollinearity
vif_data = pd.DataFrame()
vif_data["Variable"] = X_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]
print("\nVariance Inflation Factors (VIF):")
print(vif_data)
```

19. Display the $R^2$

```python
# Display R-squared and adjusted R-squared
print(f"R-squared: {model.rsquared:.4f}")
print(f"Adjusted R-squared: {model.rsquared_adj:.4f}")

# Create a visual representation of R-squared
plt.figure(figsize=(8, 6))
metrics = ['R-squared', 'Adjusted R-squared']
values = [model.rsquared, model.rsquared_adj]
colors = ['lightblue', 'lightcoral']

bars = plt.bar(metrics, values, color=colors, alpha=0.7)
plt.ylabel('Value')
plt.title('Model Goodness-of-Fit Metrics', fontsize=16, fontweight='bold')
plt.ylim(0, 1)

# Add value labels on top of bars
for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{value:.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

#Additional Enhanced Visualizations
##Residual Analysis

```python
# Residual analysis
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Residuals vs Fitted
axes[0, 0].scatter(model.fittedvalues, model.resid, alpha=0.7)
axes[0, 0].axhline(y=0, color='r', linestyle='--')
axes[0, 0].set_xlabel('Fitted Values')
axes[0, 0].set_ylabel('Residuals')
axes[0, 0].set_title('Residuals vs Fitted Values')

# Q-Q plot
sm.qqplot(model.resid, line='s', ax=axes[0, 1])
axes[0, 1].set_title('Q-Q Plot of Residuals')

# Residuals histogram
axes[1, 0].hist(model.resid, bins=10, alpha=0.7, edgecolor='black')
axes[1, 0].set_xlabel('Residuals')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].set_title('Histogram of Residuals')

# Residuals vs Independent Variables
axes[1, 1].scatter(df['Miles Traveled'], model.resid, alpha=0.7)
axes[1, 1].axhline(y=0, color='r', linestyle='--')
axes[1, 1].set_xlabel('Miles Traveled')
axes[1, 1].set_ylabel('Residuals')
axes[1, 1].set_title('Residuals vs Miles Traveled')

plt.tight_layout()
plt.show()
```

##Actual vs Predicted Values


```python
# Actual vs Predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y, model.predict(X_const), alpha=0.7, s=80)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
plt.xlabel('Actual Travel Time')
plt.ylabel('Predicted Travel Time')
plt.title('Actual vs Predicted Travel Time', fontsize=16, fontweight='bold')
plt.grid(True, alpha=0.3)

# Add R-squared text
plt.text(0.05, 0.95, f'R = {model.rsquared:.4f}', transform=plt.gca().transAxes,
         fontsize=14, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

###Feature Importance Visualization

```python
# Feature importance (using standardized coefficients)
X_standardized = (X - X.mean()) / X.std()
X_standardized_const = sm.add_constant(X_standardized)
model_standardized = sm.OLS(y, X_standardized_const).fit()

# Create coefficient plot
coef_df = pd.DataFrame({
    'Feature': X_standardized_const.columns[1:],
    'Coefficient': model_standardized.params[1:],
    'Abs_Coefficient': abs(model_standardized.params[1:])
}).sort_values('Abs_Coefficient', ascending=True)

plt.figure(figsize=(10, 6))
plt.barh(coef_df['Feature'], coef_df['Coefficient'], color='steelblue')
plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
plt.xlabel('Standardized Coefficient Value')
plt.title('Feature Importance (Standardized Coefficients)', fontsize=16, fontweight='bold')
plt.grid(True, alpha=0.3, axis='x')

# Add value labels
for i, v in enumerate(coef_df['Coefficient']):
    plt.text(v, i, f' {v:.3f}', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
